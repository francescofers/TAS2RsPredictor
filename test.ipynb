```python
from rdkit import Chem, DataStructs
import pandas as pd
import sys
%env PYENCHANT_LIBRARY_PATH=/opt/homebrew/opt/enchant/lib/libenchant-2.dylib
sys.path.append('TML/')
import Virtuous
import os

from TML_Eval import Std
```

```python
# import dataset used 
data_path = 'data'
tdf = pd.read_csv(os.path.join(data_path, 'dataset.csv'), header=0, index_col=0)
tdf.columns = tdf.columns.astype(int)

# derive parent molecules to be used for similarity search
molecules = [Virtuous.ReadMol(smi, verbose=False, type='SMILES') for smi in tdf.index]
standard = [Virtuous.Standardize(mol) for mol in molecules]
parent = [i[2] for i in standard]
parent_mol = [Chem.MolFromSmiles(i) for i in parent]

# derive inchis and fingerprints
Inchis = [Chem.MolToInchi(mol) for mol in parent_mol]
Inchis_no_stereo = [Chem.MolToInchi(mol, options='/SNon') for mol in parent_mol]
fps = [Virtuous.Calc_fps (smi) for smi in parent]

# add inchis and smiles to the dataframe
tdf.insert(0, 'Inchi', Inchis)
tdf.insert(1, 'Inchi_no_stereo', Inchis_no_stereo)
tdf.insert(2, 'Parent SMILES', parent)
tdf
```

```python
smi = 'C1CN2CC3=CCOC4CC(=O)N5C6C4C3CC2C61C7=CC=CC=C75' #strychnine
query = 'quinine' 
query = 'Cn1cnc([N+](=O)[O-])c1Sc1ncnc2[nH]cnc12'
query = 'strychnine'
query = 'quinine'
query_smiles = Chem.MolToSmiles(Virtuous.ReadMol(query, verbose=False))
query_parent = Virtuous.Standardize(Virtuous.ReadMol(query, verbose=False))[2]
query_parent_mol = Chem.MolFromSmiles(query_parent)

# derive inchi and fingerprint for query molecule
query_inchi = Chem.MolToInchi(query_parent_mol)
query_inchi_no_stereo = Chem.MolToInchi(query_parent_mol, options='/SNon')
query_fps = Virtuous.Calc_fps(query_parent)
```

```python
# check if SMILES is present
tdf[tdf['Parent SMILES'] == query_parent]
```

```python
# check if Inchi is present
tdf[tdf['Inchi'] == query_inchi]
```

```python
# check if Inchi_no_stereo is present
tdf[tdf['Inchi_no_stereo'] == query_inchi_no_stereo]
```

```python
# calculate tanimoto similarity between query and dataset
similarities = DataStructs.BulkTanimotoSimilarity(query_fps, fps)
# find index for similarity = 1.0
idx = [i for i, e in enumerate(similarities) if e > 0.5]
sims = [similarities[i] for i in idx]
if len(idx) == 0:
    print('No match found')
else: 
    fps_known = tdf.iloc[idx]

print (sims)
print (idx)
fps_known
```
